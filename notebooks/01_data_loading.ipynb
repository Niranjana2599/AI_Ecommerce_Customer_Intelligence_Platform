```python
# =========================================================
# NOTEBOOK: 01_data_loading.ipynb
# =========================================================

# =========================================================
# STEP 1 — IMPORT LIBRARIES
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Libraries Imported Successfully")
```

```python
# =========================================================
# STEP 2 — LOAD DATASETS
# =========================================================

customers_df = pd.read_csv(r'C:\Users\niran\Desktop\AI_Ecommerce_Customer_Intelligence_Platform\data\raw\customers.csv')

orders_df = pd.read_csv(r'C:\Users\niran\Desktop\AI_Ecommerce_Customer_Intelligence_Platform\data\raw\orders.csv')

order_items_df = pd.read_csv(r'C:\Users\niran\Desktop\AI_Ecommerce_Customer_Intelligence_Platform\data\raw\order_items.csv')

products_df = pd.read_csv(r'C:\Users\niran\Desktop\AI_Ecommerce_Customer_Intelligence_Platform\data\raw\products.csv')

sellers_df = pd.read_csv(r'C:\Users\niran\Desktop\AI_Ecommerce_Customer_Intelligence_Platform\data\raw\sellers.csv')

payments_df = pd.read_csv(r'C:\Users\niran\Desktop\AI_Ecommerce_Customer_Intelligence_Platform\data\raw\order_payments.csv')

reviews_df = pd.read_csv(r'C:\Users\niran\Desktop\AI_Ecommerce_Customer_Intelligence_Platform\data\raw\order_reviews.csv')

geolocation_df = pd.read_csv(r'C:\Users\niran\Desktop\AI_Ecommerce_Customer_Intelligence_Platform\data\raw\geolocation.csv')

print("All Datasets Loaded Successfully")
```

```python
# =========================================================
# STEP 3 — DATASET SHAPES
# =========================================================

print("\n================ DATASET SHAPES ================\n")

print('customers_df shape      :', customers_df.shape)
print('orders_df shape         :', orders_df.shape)
print('order_items_df shape    :', order_items_df.shape)
print('products_df shape       :', products_df.shape)
print('sellers_df shape        :', sellers_df.shape)
print('payments_df shape       :', payments_df.shape)
print('reviews_df shape        :', reviews_df.shape)
print('geolocation_df shape    :', geolocation_df.shape)
```

```python


# =========================================================
# STEP 4 — CHECK FIRST 5 ROWS
# =========================================================

print("\n================ CUSTOMERS DATA ================\n")
display(customers_df.head())

print("\n================ ORDERS DATA ================\n")
display(orders_df.head())

print("\n================ ORDER ITEMS DATA ================\n")
display(order_items_df.head())

print("\n================ PRODUCTS DATA ================\n")
display(products_df.head())

print("\n================ SELLERS DATA ================\n")
display(sellers_df.head())

print("\n================ PAYMENTS DATA ================\n")
display(payments_df.head())

print("\n================ REVIEWS DATA ================\n")
display(reviews_df.head())

print("\n================ GEOLOCATION DATA ================\n")
display(geolocation_df.head())
```

```python

# =========================================================
# STEP 5 — DATA TYPES
# =========================================================

print("\n================ DATA TYPES ================\n")

print("customers_df.dtypes:\n")
print(customers_df.dtypes)

print("\norders_df.dtypes:\n")
print(orders_df.dtypes)

print("\norder_items_df.dtypes:\n")
print(order_items_df.dtypes)

print("\nproducts_df.dtypes:\n")
print(products_df.dtypes)

print("\nsellers_df.dtypes:\n")
print(sellers_df.dtypes)

print("\npayments_df.dtypes:\n")
print(payments_df.dtypes)

print("\nreviews_df.dtypes:\n")
print(reviews_df.dtypes)

print("\ngeolocation_df.dtypes:\n")
print(geolocation_df.dtypes)
```

```python

# =========================================================
# STEP 6 — DATE CONVERSION
# =========================================================

# -----------------------------
# ORDERS DATE COLUMNS
# -----------------------------

orders_date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in orders_date_cols:
    orders_df[col] = pd.to_datetime(
        orders_df[col],
        errors='coerce'
    )


# -----------------------------
# ORDER ITEMS DATE COLUMNS
# -----------------------------

order_items_date_cols = [
    'shipping_limit_date'
]

for col in order_items_date_cols:
    order_items_df[col] = pd.to_datetime(
        order_items_df[col],
        errors='coerce'
    )


# -----------------------------
# REVIEWS DATE COLUMNS
# -----------------------------

reviews_date_cols = [
    'review_creation_date',
    'review_answer_timestamp'
]

for col in reviews_date_cols:
    reviews_df[col] = pd.to_datetime(
        reviews_df[col],
        errors='coerce'
    )

print("\nDate Columns Converted Successfully")
```

```python

# =========================================================
# STEP 7 — VERIFY UPDATED DATA TYPES
# =========================================================

print("\n================ UPDATED DATA TYPES ================\n")

print("\norders_df.dtypes:\n")
print(orders_df.dtypes)

print("\norder_items_df.dtypes:\n")
print(order_items_df.dtypes)

print("\nreviews_df.dtypes:\n")
print(reviews_df.dtypes)
```

```python

# =========================================================
# STEP 8 — CHECK MISSING VALUES
# =========================================================

print("\n================ MISSING VALUES ================\n")

datasets = {
    'customers_df': customers_df,
    'orders_df': orders_df,
    'order_items_df': order_items_df,
    'products_df': products_df,
    'sellers_df': sellers_df,
    'payments_df': payments_df,
    'reviews_df': reviews_df,
    'geolocation_df': geolocation_df
}

for name, df in datasets.items():

    print(f"\n{name}")
    print("-" * 50)

    missing = df.isnull().sum()

    missing = missing[missing > 0]

    if len(missing) == 0:
        print("No Missing Values")
    else:
        print(missing.sort_values(ascending=False))
```

```python

# =========================================================
# STEP 9 — CHECK DUPLICATES
# =========================================================

print("\n================ DUPLICATE RECORDS ================\n")

for name, df in datasets.items():

    duplicate_count = df.duplicated().sum()

    print(f"{name} --> Duplicate Rows: {duplicate_count}")
```

```python

# =========================================================
# STEP 10 — BASIC STATISTICAL SUMMARY
# =========================================================

print("\n================ NUMERICAL SUMMARY ================\n")

display(customers_df.describe())

display(order_items_df.describe())

display(products_df.describe())

display(payments_df.describe())

display(reviews_df.describe())
```

```python

# =========================================================
# STEP 11 — BASIC BUSINESS KPIs
# =========================================================

print("\n================ BUSINESS KPIs ================\n")

# Total Customers
total_customers = customers_df['customer_unique_id'].nunique()

# Total Orders
total_orders = orders_df['order_id'].nunique()

# Total Products
total_products = products_df['product_id'].nunique()

# Total Sellers
total_sellers = sellers_df['seller_id'].nunique()

# Total Revenue
total_revenue = payments_df['payment_value'].sum()

# Average Order Value
avg_order_value = payments_df['payment_value'].mean()

# Delivered Orders
delivered_orders = (
    orders_df['order_status'] == 'delivered'
).sum()

# Cancelled Orders
cancelled_orders = (
    orders_df['order_status'] == 'canceled'
).sum()

# Delivery Success Rate
delivery_success_rate = (
    delivered_orders / total_orders
) * 100

# Average Review Score
avg_review_score = reviews_df['review_score'].mean()

print(f"Total Customers           : {total_customers:,}")
print(f"Total Orders              : {total_orders:,}")
print(f"Total Products            : {total_products:,}")
print(f"Total Sellers             : {total_sellers:,}")
print(f"Total Revenue             : ${total_revenue:,.2f}")
print(f"Average Order Value       : ${avg_order_value:.2f}")
print(f"Delivered Orders          : {delivered_orders:,}")
print(f"Cancelled Orders          : {cancelled_orders:,}")
print(f"Delivery Success Rate     : {delivery_success_rate:.2f}%")
print(f"Average Review Score      : {avg_review_score:.2f}")
```

```python

# =========================================================
# STEP 12 — CREATE TIME FEATURES
# =========================================================

# Delivery Duration
orders_df['delivery_duration_days'] = (
    orders_df['order_delivered_customer_date']
    - orders_df['order_purchase_timestamp']
).dt.days

# Delivery Delay
orders_df['delivery_delay_days'] = (
    orders_df['order_delivered_customer_date']
    - orders_df['order_estimated_delivery_date']
).dt.days

# Approval Time
orders_df['approval_time_hours'] = (
    orders_df['order_approved_at']
    - orders_df['order_purchase_timestamp']
).dt.total_seconds() / 3600

# Purchase Features
orders_df['purchase_year'] = (
    orders_df['order_purchase_timestamp'].dt.year
)

orders_df['purchase_month'] = (
    orders_df['order_purchase_timestamp'].dt.month
)

orders_df['purchase_day'] = (
    orders_df['order_purchase_timestamp'].dt.day
)

orders_df['purchase_weekday'] = (
    orders_df['order_purchase_timestamp'].dt.day_name()
)

print("\nTime Features Created Successfully")
```

```python


# =========================================================
# STEP 13 — CHECK NEW FEATURES
# =========================================================

print("\n================ NEW TIME FEATURES ================\n")

display(
    orders_df[
        [
            'delivery_duration_days',
            'delivery_delay_days',
            'approval_time_hours',
            'purchase_year',
            'purchase_month',
            'purchase_day',
            'purchase_weekday'
        ]
    ].head()
)
```

```python

# =========================================================
# STEP 14 — UNDERSTAND TABLE RELATIONSHIPS
# =========================================================

print("\n================ TABLE RELATIONSHIPS ================\n")

print("""
customers_df
    └── customer_id
            ↓
orders_df
    └── order_id
            ↓
order_items_df
        ├── product_id → products_df
        └── seller_id  → sellers_df

orders_df
    ├── order_id → payments_df
    └── order_id → reviews_df
""")
```
